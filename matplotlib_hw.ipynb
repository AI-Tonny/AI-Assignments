### Завдання 1: Лінійний графік "Трекінг навчання"
Створи графік, який відображає твій прогрес у вивченні технологій за тиждень.

1. Створи масив `days` (дні тижня) та масив `hours` (кількість годин, витрачених на навчання).
2. Побудуй лінійний графік.
3. **Вимоги до оформлення:**
    * Колір лінії — зелений.
    * Додай маркери (точки) на кожному дні.
    * Підпиши осі та дай назву графіку.
    * Увімкни сітку.

```python
import matplotlib.pyplot as plt

days = ['Пн', 'Вв', 'Ср', 'Чт', 'Пт', 'Сб', 'Нд']
hours = [2, 3, 1, 4, 2, 5, 3]

plt.figure(figsize=(8, 5))
plt.plot(days, hours, color='green', marker='o')

plt.xlabel('Дні тижня')
plt.ylabel('Кількість годин')
plt.title('Трекінг навчання за тиждень')
plt.grid(True)

plt.show()
```

### Завдання 2: Робота з декількома вікнами (Subplots)
Тобі потрібно порівняти дві математичні функції на одному полотні, але в різних зонах.

1. Створи полотно (`Figure`) з двома графіками поруч (1 рядок, 2 стовпці).
2. На першому графіку зобрази функцію $y = x^2$.
3. На другому графіку зобрази функцію $y = x^3$
4. Використай для них різні стилі ліній (наприклад, суцільна та пунктирна).
5. **Бонус:** Використай команду `plt.style.use()`, щоб змінити загальну тему (наприклад, на `'seaborn-v0_8'` або `'bmh'`).

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8')

x = np.linspace(-10, 10, 100)
y1 = x**2
y2 = x**3

fix, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(x, y1, linestyle='-', color='blue')
ax[0].set_title('$y = x^2$')
ax[0].grid(True)

ax[1].plot(x, y2, linestyle='--', color='red')
ax[1].set_title('$y = x^3$')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

### Завдання 3: Аналіз випадкових величин (Гістограма)
Це завдання на розуміння розподілу даних.

1. Згенеруй 1000 випадкових чисел за допомогою `np.random.randn(1000)` (нормальний розподіл).
2. Побудуй гістограму (`plt.hist`) для цих чисел.
3. Встанови кількість "кошиків" (bins) рівною 30.
4. Зміни колір стовпчиків на будь-який інший (наприклад, `'orange'`) і додай прозорість (`alpha=0.7`).

---

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='orange', alpha=0.7, edgecolor='black')

plt.title('Гістограма нормального розподілу')
plt.xlabel('Значення')
plt.ylabel('Частота')
plt.grid(axis='y', alpha=0.3)

plt.savefig('histogram.png')

plt.show()
```
